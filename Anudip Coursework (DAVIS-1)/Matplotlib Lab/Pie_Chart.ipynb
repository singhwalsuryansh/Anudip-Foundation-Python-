```python
# PIE CHART

# Q31 — Market Share of Smartphone Brands 

# Import matplotlib library
import matplotlib.pyplot as plt

# Smartphone brand market share
brands = ["Apple", "Samsung", "Xiaomi", "Oppo", "Others"]

# Market share percentages
share = [30, 25, 20, 15, 10]

# Create chart size
plt.figure(figsize=(8,6))

# Create pie chart
plt.pie(share, labels=brands, autopct="%1.1f%%")

# Chart title
plt.title("Smartphone Market Share")

# Display chart
plt.show()
```

```python
# Q32 — Monthly Expense Distribution

# Import matplotlib
import matplotlib.pyplot as plt

# Expense categories
categories = ["Food", "Rent", "Travel", "Shopping", "Bills"]

# Amount spent
amount = [5000, 12000, 3000, 4000, 2500]

# Chart size
plt.figure(figsize=(8,6))

# Pie chart
plt.pie(amount, labels=categories, autopct="%1.1f%%")

# Title
plt.title("Monthly Expense Distribution")

# Show chart
plt.show()
```
