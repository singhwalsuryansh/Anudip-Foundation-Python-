```python
# HISTOGRAM PLOT

# Q36 — Student Marks Distribution

# Import matplotlib
import matplotlib.pyplot as plt

# Student marks dataset
marks = [45,50,55,60,65,70,72,75,78,80,82,85,88,90,92,95]

# Chart size
plt.figure(figsize=(8,5))

# Histogram
plt.hist(marks, bins=5)

# Title
plt.title("Student Marks Distribution")

# Axis labels
plt.xlabel("Marks")
plt.ylabel("Number of Students")

# Grid
plt.grid(True)

# Display chart
plt.show()
```

```python
# Q37 — Age Distribution

# Import matplotlib
import matplotlib.pyplot as plt

# Age dataset
age = [18,20,21,22,23,25,26,27,29,30,32,34,35,36,38,40]

# Chart size
plt.figure(figsize=(8,5))

# Histogram
plt.hist(age, bins=6)

# Title
plt.title("Age Distribution")

# Axis labels
plt.xlabel("Age")
plt.ylabel("Frequency")

# Grid
plt.grid(True)

# Show chart
plt.show()
```
