```python
# BAR CHART

# 11. Create a Bar Chart showing students in different courses.

# Import matplotlib
import matplotlib.pyplot as plt

# Dataset
courses = ["Python","Java","C++","Web","AI"]
students = [40,35,25,30,20]

# Create bar chart
plt.figure(figsize=(8,5))
plt.bar(courses, students, color="skyblue", edgecolor="black")

# Add data values on bars
for i in range(len(courses)):
    plt.text(i, students[i] + 1, students[i], ha='center')

# Labels and title
plt.title("Students Enrolled in Courses")
plt.xlabel("Courses")
plt.ylabel("Number of Students")
plt.grid(axis='y')

plt.show()
```

```python
#12. Plot Sales by Product.

# Step 1: Store the dataset
products = ["Laptop", "Mobile", "Tablet", "Keyboard", "Mouse"]
sales = [120, 200, 90, 150, 180]

# Step 2: Create a bar chart
plt.figure(figsize=(8,5))  # Set chart size

plt.bar(products, sales, color='skyblue', edgecolor='black')

# Step 3: Add data values on top of each bar
for i in range(len(products)):
    plt.text(i, sales[i] + 3, sales[i], ha='center')

# Step 4: Add chart title and axis labels
plt.title("Product Sales Bar Chart")
plt.xlabel("Products")
plt.ylabel("Sales")

# Step 5: Add grid for better visualization
plt.grid(axis='y')

# Step 6: Display the chart
plt.show()
```
